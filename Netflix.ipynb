```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df= pd.read_csv('Netflix_Userbase.csv')
df.head(20)
df.shape 
df.info()
df.nunique()
df['Country'].unique()
df['Subscription Type'].unique()
df.isnull().sum()
df.describe() 
df.describe(include='object')
df.describe(include='all')
df.duplicated(keep='first').sum()
df['Subscription Type'].value_counts()
df['Country'].value_counts()
df['Gender'].value_counts()
df['Device'].value_counts()
df['Country'].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=90, shadow=False, legend=False);
df['Gender'].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=90, shadow=False, legend=False);
df['Subscription Type'].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=90, shadow=False, legend=False);
df['Gender'].value_counts().plot(kind='bar')
df['Subscription Type'].value_counts().plot(kind='bar')
df['Country'].value_counts().plot(kind='bar')
df.drop('User ID', axis=1, inplace=True)
filtre = ['United Kingdom', 'Germany', 'France', 'Spain', 'Italy']
df = df[df['Country'].isin(filtre)]
df['Country'].unique()
df['Country'].value_counts()
pd.DataFrame({  'Nombre': df['Gender'].value_counts(), 
               'Pourcentage (%)': df['Gender'].value_counts(normalize=True)
               })


pd.DataFrame({  'Nombre': df['Age'].value_counts(), 
                'Pourcentage (%)': df['Age'].value_counts(normalize=True)
               })


pd.DataFrame({  'Nombre': df['Device'].value_counts(), 
                'Pourcentage (%)': df['Device'].value_counts(normalize=True)
                })




pd.crosstab(df['Subscription Type'], df['Gender'])
pd.crosstab(df['Subscription Type'], df['Gender'], normalize='index') 
pd.crosstab(df['Subscription Type'], df['Gender'], normalize='columns') 
pd.crosstab(df['Subscription Type'], df['Gender'], normalize='all') 

pd.pivot_table(df, index='Country', columns=['Gender','Device'], values='Subscription Type', aggfunc='count')

```
